```python
#!/usr/bin/env python
# coding: utf-8

# In[1]:


# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.mixture import GaussianMixture
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split


# In[ ]:





# In[2]:


# Load the dataset
mall_customers_df = pd.read_csv('Mall_Customers.csv')


# In[28]:


# Display the column names of the dataset
print(mall_customers_df.columns)


# In[3]:


# Display the first few rows of the dataset
print(mall_customers_df.head())

# Check for any missing values
print(mall_customers_df.isnull().sum())


# In[4]:


# Summary statistics
print(mall_customers_df.describe())


# In[29]:


# Statistical Graph Quality (Box Plot): Age Distribution
plt.figure(figsize=(8, 6))
sns.boxplot(x='Age', data=mall_customers_df, palette='Set1')
plt.title('Age Distribution')
plt.show()


# In[32]:


# Categorical Graph Quality (Pie Chart): Gender Distribution
plt.figure(figsize=(8, 6))
mall_customers_df['Genre'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'pink'])
plt.title('Genre Distribution')
plt.ylabel('')
plt.show()


# In[30]:


# Relational Graph Quality (Scatter Plot): Spending Score vs Annual Income
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', data=mall_customers_df, hue='Genre', palette='Set1')
plt.title('Spending Score vs Annual Income')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()



# In[9]:


# Data preprocessing
X = mall_customers_df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


# In[10]:



# Dimensionality reduction with PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)


# In[11]:


# Hierarchical Clustering
plt.figure(figsize=(12, 8))
dendrogram(linkage(X_scaled, method='ward'))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distance')
plt.show()


# In[37]:


from sklearn.cluster import DBSCAN

# Clustering (DBSCAN)
dbscan = DBSCAN(eps=0.5, min_samples=5)
mall_customers_df['cluster_dbscan'] = dbscan.fit_predict(X_scaled)

# Number of clusters in DBSCAN, ignoring noise if present.
n_clusters_dbscan = len(set(mall_customers_df['cluster_dbscan'])) - (1 if -1 in mall_customers_df['cluster_dbscan'] else 0)
print("Number of clusters in DBSCAN:", n_clusters_dbscan)

# Scatter plot of PCA components with DBSCAN clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=mall_customers_df['cluster_dbscan'], palette='Set1')
plt.title('PCA Components with DBSCAN Clusters')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.show()


# In[33]:


# Pair plot for numerical variables
plt.figure(figsize=(12, 8))
sns.pairplot(mall_customers_df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)', 'Genre']], hue='Genre', palette='Set1')
plt.suptitle('Pair Plot of Numerical Variables', y=1.02)
plt.show()


# In[34]:


# Correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(mall_customers_df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()


# In[35]:


# Histograms for numerical variables
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
sns.histplot(mall_customers_df['Age'], bins=20, kde=True, color='skyblue')
plt.title('Age Distribution')

plt.subplot(1, 3, 2)
sns.histplot(mall_customers_df['Annual Income (k$)'], bins=20, kde=True, color='salmon')
plt.title('Annual Income Distribution')

plt.subplot(1, 3, 3)
sns.histplot(mall_customers_df['Spending Score (1-100)'], bins=20, kde=True, color='green')
plt.title('Spending Score Distribution')

plt.tight_layout()
plt.show()


# In[36]:


# Clustering (K-Means)
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(X_scaled)
mall_customers_df['cluster'] = kmeans.labels_


# In[13]:



# Silhouette score
silhouette_avg = silhouette_score(X_scaled, kmeans.labels_)
print("Silhouette Score (K-Means):", silhouette_avg)


# In[14]:


# Clustering (Gaussian Mixture Models)
gmm = GaussianMixture(n_components=5, random_state=42)
gmm.fit(X_scaled)
mall_customers_df['cluster_gmm'] = gmm.predict(X_scaled)


# In[15]:


# Scatter plot of PCA components with GMM clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=mall_customers_df['cluster_gmm'], palette='Set1')
plt.title('PCA Components with GMM Clusters')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.show()


# In[16]:


# Fitting (Polynomial Regression)
X_train, X_test, y_train, y_test = train_test_split(X[['Age']], X['Spending Score (1-100)'], test_size=0.2, random_state=42)


# In[17]:


# Create polynomial features
poly_features = PolynomialFeatures(degree=2)
X_poly = poly_features.fit_transform(X_train)


# In[18]:


# Fit polynomial regression model
poly_reg = LinearRegression()
poly_reg.fit(X_poly, y_train)


# In[19]:


# Predict
X_pred = np.linspace(X_train['Age'].min(), X_train['Age'].max(), 100).reshape(-1, 1)
X_pred_poly = poly_features.transform(X_pred)
y_pred = poly_reg.predict(X_pred_poly)


# In[20]:


# Scatter plot with polynomial regression curve
plt.figure(figsize=(8, 6))
plt.scatter(X_test['Age'], y_test, label='Actual Data')
plt.plot(X_pred, y_pred, color='red', label='Polynomial Regression Curve')
plt.title('Age vs Spending Score with Polynomial Regression')
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()


# In[21]:


# Calculate Mean Squared Error
mse = mean_squared_error(y_test, poly_reg.predict(poly_features.transform(X_test)))
print("Mean Squared Error (Polynomial Regression):", mse)


# In[22]:


# Fitting (Support Vector Regression)
svr_model = SVR(kernel='linear')
svr_model.fit(X_train, y_train)


# In[23]:


# Predict
y_pred_svr = svr_model.predict(X_test)


# In[24]:


# Scatter plot with SVR line
plt.figure(figsize=(8, 6))
plt.scatter(X_test['Age'], y_test, label='Actual Data')
plt.plot(X_test['Age'], y_pred_svr, color='red', label='SVR Prediction')
plt.title('Age vs Spending Score with SVR')
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()


# In[25]:



# Calculate R-squared score
r2 = r2_score(y_test, y_pred_svr)
print("R-squared Score (SVR):", r2)


# # Calculate mean, median, standard deviation, skewness, and kurtosis
mean_values = mall_customers_df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].mean()
median_values = mall_customers_df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].median()
std_deviation = mall_customers_df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].std()
skewness = mall_customers_df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].skew()
kurtosis = mall_customers_df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].kurtosis()

print("Mean Values:")
print(mean_values)
print("\nMedian Values:")
print(median_values)
print("\nStandard Deviation:")
print(std_deviation)
print("\nSkewness:")
print(skewness)
print("\nKurtosis:")
print(kurtosis)

# Calculate correlation matrix
correlation_matrix = mall_customers_df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].corr()

print("\nCorrelation Matrix:")
print(correlation_matrix)





```
